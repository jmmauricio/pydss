```python
@numba.jit(nopython=True)
def run_eval():
    N_steps = int(1e3)
    N_states = 1
    x = np.zeros((N_states,1))+4
    f = np.zeros((N_states,1))
    X = np.zeros((N_states,N_steps))
    out = np.zeros((3,N_steps))
    #self.out_cplx = np.zeros((72,N_steps),np.complex128)
        
    T = np.zeros((N_steps,1))
    
    obj_w = mech.wind_turbine()
    obj_w.nu = 8.0
    obj_w.tau_r = 100
    dt = 5.0e-3
    t = 0.0
    X[:,0] = x[:,0]
    T[0,0] = t
    for it in range(1,N_steps):  
            t += dt
        
            obj_w.x = x

          
            obj_w.f_eval()
            f[0] = obj_w.f

            x = x + dt*f
            
            X[:,it] = x[:,0]
            out[0,it] =obj_w.p_m
            T[it,0] = t
            
    return T,X,out
```

```python
T,X,out = run_eval()
```

```python
fig = figure(width=600, height=300)
decim = 1
fig.line(T[0:-1:decim,0],X[0,0:-1:decim], line_width=1, color='red')
show(fig)

fig = figure(width=600, height=300)
decim = 1
fig.line(T[0:-1:decim,0],out[0,0:-1:decim], line_width=1, color='red')
show(fig)
```
