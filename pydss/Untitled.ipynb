```python
import numpy as np
import xmlrpc.client

from bokeh.io import output_notebook, show
from bokeh.plotting import figure
from bokeh.layouts import gridplot
from bokeh.models import ColumnDataSource, HoverTool, CustomJS
output_notebook()
```

```python
pro  =xmlrpc.client.ServerProxy("http://localhost:1080/RPC2")  
```

```python
sys = pro.plecs.load("/home/jmmauricio/Documents/public/jmmauricio6/RESEARCH/thermal/Temp_OneLeg_2.plecs")
```

```python
w = 2.0*np.pi*50.0
V_s = 400.0/np.sqrt(3);
V_s_peak = V_s*np.sqrt(2);

for i_value in np.linspace(-100,100,10):
    
    I_l = i_value + 1j*0.0; # rms

    R= 0.1
    L= 0.005
    Z = R+1j*w*L;
    E = V_s_peak - Z*I_l*np.sqrt(2);

    E_peak = np.abs(E);
    E_angle = np.angle(E);

    opts = {'ModelVars' :  {'E_peak' : float(E_peak),
                            'E_angle' : float(E_angle),
                            'V_s_peak':float(V_s_peak),
                            'R' : float(R),
                            'L':float(L),
                           } } 


    #%timeit pro.plecs.simulate("Temp_OneLeg_2", opts) 
    result = pro.plecs.simulate("Temp_OneLeg_2", opts) 
    
    t  = result['Time']
    i_l = result['Values'][0]
    t_rms = np.linspace(0.98,1.0,1000)
    i_new = np.interp(t_rms,t,i_l)
    rms = np.sqrt(np.mean(np.square(i_new)))
    print(E_peak,rms)
```
